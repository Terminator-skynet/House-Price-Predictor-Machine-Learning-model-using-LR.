# Summary
### Summary of House Price Prediction Findings

My Random Forest model revealed that **median income** and **geographic location** (proximity to the ocean and latitude/longitude) are the strongest drivers of California house prices. In plain terms, the model achieves an R-squared score of around 80–82%, meaning it can accurately explain roughly 80% of the variance in home values, making it highly reliable for market estimates. A surprising insight from the data was that population density and total rooms had a distorted, skewed impact until we applied log transformations, showing that massive developments don't automatically equal higher luxury value.

Based on these findings, we recommend that real estate businesses prioritize investments in high-income coastal pockets and use engineered metrics like the bedroom-to-room ratio to identify underpriced, high-potential properties.
